###  Functional  Model Building

```python
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.layers import Input, Flatten, Dense
```

```python
Xtrain.shape
```

```python
feed = Input(shape=(28,28))
```

```python
flat = Flatten()(feed)
```

```python
dh1 = Dense(32, activation = "relu")(flat)
```

```python
dh2 = Dense(64, activation = 'relu')(dh1)
```

```python
dh3 = Dense(32, activation = 'relu')(dh2)
```

```python
out = Dense(10, activation = 'softmax')(dh3)
```

```python
model = Model(inputs = feed, outputs = out )
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
```

```python
model.fit(Xtrain_scale, ytrain, epochs= 20)
```

```python
model.evaluate(Xtest_scale,ytest)
```
